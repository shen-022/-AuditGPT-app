```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
original_data = pd.read_csv("FFUT_FDT.csv")
original_data.head(10)
```

```python
cleaned_data = original_data.copy()
cleaned_data = cleaned_data[cleaned_data['Agmtcd'].str.startswith('IH')]
cleaned_data
```

```python
#mean_ann_return 
#For now we assume that we put up the entire current notional exposure value as capital
cleaned_data['IH_daily_rr'] = (cleaned_data['Stprc'] - cleaned_data['Ystprc']) / cleaned_data['Ystprc']
cleaned_data
```

```python
cleaned_data = cleaned_data.loc[cleaned_data.groupby('Trddt')['Volume'].idxmax()]
cleaned_data
```

```python
daily_rr = cleaned_data.groupby('Trddt')['IH_daily_rr'].mean()
daily_rr
```

```python
daily_avg_return = daily_rr.mean()
daily_avg_return
```

```python
mean_ann_return = daily_avg_return * 252
mean_ann_return
```

```python
#Average_drawdowm
cleaned_data["IH_daily_drawdowm"] = (cleaned_data['Hiprc'] - cleaned_data['Loprc'])/cleaned_data['Hiprc']
cleaned_data
```

```python
Average_drawdowm = cleaned_data["IH_daily_drawdowm"].mean()
Average_drawdowm
```

```python
#Ann std
daily_std = daily_rr.std()
daily_std
```

```python
Ann_std = daily_std * np.sqrt(252)
Ann_std
```

```python
#SR  是用annual算 不是daily算
SR = (mean_ann_return/Ann_std)
SR
```

```python
#SKEW
skewness = daily_rr.skew()
skewness
```

```python
#lower&upper tail, lower&upper percentile ratio 2010
df_2015 = cleaned_data.copy()
df_2015 = cleaned_data[cleaned_data['Trddt'].str.startswith('2015')]
df_2015
```

```python
df_2015["IH_daily_rr"] = df_2015["IH_daily_rr"]- daily_avg_return
df_2015 = df_2015.reset_index()
df_2015
```

```python
data = np.arange(0,176,1)
p30 = int(np.percentile(data, 30))
p70 =int(np.percentile(data, 70))
p1 = int(np.percentile(data, 1))
p99 =int(np.percentile(data, 99))
print(p30,p70,p1,p99)
```

```python
left_value = df_2015.iloc[p30, -2]
right_value = df_2015.iloc[p70, -2]
value1st = df_2015.iloc[p1, -2]
value99st = df_2015.iloc[p99,-2]
```

```python
lower_pr = value1st / left_value
upper_pr = value99st / right_value
print(lower_pr,upper_pr)
```

```python
lower_tail = lower_pr/4.43
upper_tail = upper_pr/4.43
print(lower_tail,upper_tail)
```

```python
#Account curve (cumulated sum of percentage returns)
cumulative_return = (1 + daily_rr).cumprod() - 1
cumulative_return
```

```python
df = pd.DataFrame(cumulative_return)
df = df.reset_index()
df
```

```python
plt.figure(figsize=(15, 8))
plt.plot(df['Trddt'], df['IH_daily_rr'],'b-')
plt.title("Cumulative Percentage Returns")
plt.ylabel("Cumulative Return")
plt.xlabel("Time Period")
plt.xticks(df['Trddt'][::200], rotation=45)
plt.tight_layout()
plt.grid(True)
plt.legend()
plt.show()
```
